```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from sklearn.cluster import KMeans

I = Image.open("Lago_Victoria.jpg")

plt.figure(figsize=(8,8))
plt.imshow(I)
plt.axis('off')
plt.show()
```

```python
I1 = I.convert('L')
a = np.asarray(I1,dtype=np.float)
print(a.shape)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(a,cmap='gray')
plt.axis('off')
plt.show()
```

```python
x , y = a.shape
print('Dimensiones de a ', a.shape)

a1 = a.reshape(x*y,1)
print('Dimensiones de a1 ', a1.shape)
```

```python
print(a1[0])
```

```python
k_means = KMeans(n_clusters=3)
k_means.fit(a1)
```

```python
centroides = k_means.cluster_centers_
etiquetas = k_means.labels_

print(centroides.shape)
print(centroides[0])
print(centroides[1])
print(centroides[2])

print(etiquetas.shape)
print(etiquetas[0])
print(etiquetas[100])
print(etiquetas[10000])
print(etiquetas[200000])
print(etiquetas[150000])
```

```python
a2 = centroides[etiquetas]
print('Dimensiones de a2 ', a2.shape)
print(a2[0])
```

```python
a3 = a2.reshape(x,y)
print('Dimensiones de a3 ', a3.shape)

plt.figure(figsize=(8,8))
plt.imshow(a3,cmap='gray')
plt.axis('off')
plt.show()
```
